```python
import csv
import sqlite3
import pandas as pd
import os
```

```python
def convert_sqlite_to_csv(inputFolder, ext, tableName):
    """ inputFolder - Folder where sqlite files are located. 
        ext - Extension of your sqlite file (eg. db, sqlite, sqlite3 etc.)
        tableName - table name from which you want to select the data.
    """
    csvWriter = csv.writer(open(inputFolder+'/{}.csv'.format(tableName), 'w', newline=''))
    for file1 in os.listdir(inputFolder):
        if file1.endswith('.'+ext):
            conn = sqlite3.connect(inputFolder+'/'+file1)
            cursor = conn.cursor()
            cursor.execute("SELECT * FROM "+tableName)
            rows = cursor.fetchall()
            table_info = cursor.execute("PRAGMA table_info({})".format(tableName)).fetchall()
            column_names = [elt[1] for elt in table_info]
            print(table_info)
            csvWriter.writerow(column_names)
            for row in rows:
                csvWriter.writerow(row)
            continue
        else:
            continue
convert_sqlite_to_csv("data", "db", "people")
```
